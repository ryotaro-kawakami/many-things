```
from keras.models import model_from_json
from keras.utils import np_utils
import numpy as np
from keras.preprocessing.image import img_to_array, load_img
import cv2
from keras_applications.vgg16 import preprocess_input
```

```
model1 = model_from_json(open('/content/drive/My Drive/stanford_cars_dataset/11-18VGGexperiment_model.json').read())
model2 = model_from_json(open('/content/drive/My Drive/stanford_cars_dataset/11-18VGGexperiment_model.json').read())
model3 = model_from_json(open('/content/drive/My Drive/stanford_cars_dataset/11-18VGGexperiment_model.json').read())
model4 = model_from_json(open('/content/drive/My Drive/stanford_cars_dataset/11-18VGGexperiment_model.json').read())
model5 = model_from_json(open('/content/drive/My Drive/stanford_cars_dataset/11-18VGGexperiment_model.json').read())
model1.load_weights('/content/drive/My Drive/stanford_cars_dataset/11-18-1_15.h5')
model2.load_weights('/content/drive/My Drive/stanford_cars_dataset/11-18-2_13.h5')
model3.load_weights('/content/drive/My Drive/stanford_cars_dataset/11-18-3_13.h5')
model4.load_weights('/content/drive/My Drive/stanford_cars_dataset/11-18-4_12.h5')
model5.load_weights('/content/drive/My Drive/stanford_cars_dataset/11-18-5_11.h5')
```

```
img_path = '/content/drive/My Drive/stylegan/RyotarobeautyTrue.jpeg' #判定したい画像のpathを入力
img = img_to_array(load_img(img_path, target_size=(224,224)))
img_nad = img_to_array(img)
img_nad = np.expand_dims(img_nad, axis=0)
```

```
import matplotlib.pyplot as plt
image = cv2.imread(img_path)
plt.imshow(image)
```

```
image_224 = cv2.resize(image, (224,224))
image_224 = np.expand_dims(image_224,axis=0)
image_224.shape
```

```
pred1 = model1.predict(image_224, batch_size=1, verbose=0)
pred2 = model2.predict(image_224, batch_size=1, verbose=0)
pred3 = model3.predict(image_224, batch_size=1, verbose=0)
pred4 = model4.predict(image_224, batch_size=1, verbose=0)
pred5 = model5.predict(image_224, batch_size=1, verbose=0)
score1 = np.max(pred1)
score2 = np.max(pred2)
score3 = np.max(pred3)
score4 = np.max(pred4)
score5 = np.max(pred5)
pred_label1 = np.argmax(pred1)
pred_label2 = np.argmax(pred2)
pred_label3 = np.argmax(pred3)
pred_label4 = np.argmax(pred4)
pred_label5 = np.argmax(pred5)
print('number',pred_label1)
print('score',score1)
print('number',pred_label2)
print('score',score2)
print('number',pred_label2)
print('score',score3)
print('number',pred_label3)
print('score',score3)
print('number',pred_label4)
print('score',score4)
print('number',pred_label5)
print('score',score5)
```

```
truepred=(pred1+pred2+pred3+pred4+pred5)/5
truescore=(score1+score2+score3+score4+score5)
truepred_label = np.argmax(truepred)
print('number',truepred_label)
print('score',score5)
```
